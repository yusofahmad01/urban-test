```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.set_style('darkgrid')
f, axes = plt.subplots(2,1, figsize=(7,7), sharex=True)
sns.distplot(df['tip'], color='blue', ax=axes[0], kde=False)
sns.distplot(df['total_bill'], color='olive', ax=axes[1], kde=False)
```

```python

g=sns.FacetGrid(df, row= 'smoker', height =3, aspect=2) #height of each facet

g=g.map(plt.hist, 'total_bill')
plt.show()
```
